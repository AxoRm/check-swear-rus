# Введение

В последние годы значительно возросла потребность в автоматической фильтрации нецензурной лексики в интернет-коммуникациях, особенно в образовательных учреждениях, таких как школы и университеты. Среди методов, применяемых для этой цели, выделяются линейные классификаторы и нейронные сети. Каждый из этих подходов имеет свои преимущества и недостатки, важные для понимания при выборе конкретной технологии для внедрения.

## Линейная классификация мата

Линейные классификаторы работают на основе разделения данных линейной границей, что делает их относительно простыми для понимания и реализации. Преимуществами линейной классификации являются высокая скорость работы и низкие требования к вычислительным ресурсам, что делает их идеальными для использования в условиях ограниченных ресурсов.

### Преимущества:
- **Высокая скорость обработки**: Линейные классификаторы могут быстро обрабатывать большие объемы текстовых данных.
- **Простота реализации**: Линейные модели легче разрабатывать и внедрять, поскольку они не требуют сложной архитектуры.

### Недостатки:
- **Ограниченная точность**: Линейные классификаторы могут не справляться с высокой степенью сложности задачи из-за ограниченности линейной разделяющей границы.
- **Неадекватность для сложных зависимостей**: Они менее эффективны в случаях, когда необходимо учитывать контекст и глубинные семантические связи в тексте.

## Нейросетевые подходы

Нейронные сети, включая глубокое обучение, предоставляют более сложные и мощные средства для классификации текстов, в том числе для выявления нецензурной лексики. Они способны улавливать сложные зависимости и контекстуальные значения слов, что значительно повышает точность классификации.

### Преимущества:
- **Высокая точность**: Способность к обучению на больших объемах данных и выявлению сложных зависимостей обеспечивает высокую точность классификации.
- **Гибкость**: Нейронные сети могут адаптироваться к разнообразным задачам и условиям использования за счет обучения на соответствующих данных.

### Недостатки:
- **Высокие требования к ресурсам**: Обучение и использование нейронных сетей требует значительных вычислительных ресурсов.

### Ожидания

Технически, нецензурность комментария определяется наличием нецензурных слов в нем, которых, так или иначе, все же конечное количество. Набрав достаточное количество данных, в пространстве слов (_bag-of-words_) мы можем ожидать достаточно точное выделение подпространства этих самых слов, которое будет неплохо разделятся простым линейным классификатором. А один из лучших способов это проверить - непосредственно обучить и протестировать модель.

```python
csv_directory = './cleaned_data'
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]

dataframes = []

for file in csv_files:
    df = pd.read_csv("./cleaned_data/" + file)
    dataframes.append(df)
    
df = pd.concat(dataframes)
```

# Создание датасета для классификации

Для обучения модели классификации нецензурной лексики был сформирован датасет, состоящий из около 800 000 комментариев, собранных из различных школьных и студенческих чатов. Создание качественного датасета является ключевым шагом в подготовке к обучению модели, так как от качества исходных данных напрямую зависят точность и эффективность классификации.

## Препроцессинг данных

Перед включением в датасет, каждый комментарий подвергся тщательной предобработке с целью унификации данных и улучшения качества обучения модели. Препроцессинг включал в себя следующие шаги:

- **Приведение к нижнему регистру**: Все текстовые данные были приведены к нижнему регистру для исключения дублирования слов, отличающихся только регистром букв.
- **Замена английских букв на русские**: Для предотвращения путаницы и сокращения размерности пространства признаков, английские буквы, используемые в качестве замен русских, были заменены на соответствующие русские буквы.
- **Удаление ненужных символов**: Из текста удалялись специальные символы, знаки препинания и числа, не несущие смысловой нагрузки для задачи классификации.

Эти шаги препроцессинга были предприняты для упрощения последующего анализа текстов и сокращения размерности пространства признаков в модели "мешок слов" (bag of words). В классе препроцессинг есть многие другие функции, включая удаление long vowels и слияния всех слов в одну строку. Это сделано для двухуровневой проверки регулярными выражениями. 

## Важность предобработки данных

Тщательный препроцессинг текстовых данных играет важную роль в подготовке к эффективному обучению модели классификации. Он позволяет не только улучшить качество модели, но и существенно сократить время обучения за счет уменьшения размерности данных и исключения избыточной информации. Несмотря на то, что SVM достаточно хорошо работает на данных большой размерности, различать слова _сделал_ и _сделала_ модели не следует, хотя бы из соображений экономии памяти.

## Обучение baseline модели

```python
from sklearn.model_selection import train_test_split

X, y = df.messages, df.is_profane
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED, stratify=df.is_profane)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X_vec = vectorizer.fit_transform(X_train)
X_vec_test = vectorizer.transform(X_test)
```

### Тюнинг

Для тюнинга используем обычный `GridSearchCV`. И регуляризация `C`, и `class_weight` - это параметры, сильно влияющие на вид разделяющей гиперплоскости. Из-за имбаланса в датасете логично будет проверить два режима для выставления весов, параметр регуларизации будет разрешать нам ошибаться на выборке. Конечно, лучшей оценкой для такой задачи будет _f1 score_, так как с одной стороны нам важно нахождение всех матных комментариев, с другой - мисклассификация нематного комментария тоже нежелательна. Но лично для себя хороший _recall_ я выделил как чуть более важную цель, так как пропустить мат в моих задачах было очень нежелательно.

```python
# tuning SVM model
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# setting several C parameters for margins
# dual is set to False as n_samples > n_features

param_grid = {'C' : np.array([0.001, 0.01, 0.1, 1., 10., 100.]),
              'class_weight': [None, 'balanced']}

scoring = {
    'f1': 'f1',
    'recall': 'recall',
    'precision': 'precision',
    'accuracy': 'accuracy',
    'balanced_accuracy': 'balanced_accuracy'
}

grid_clf_cv = GridSearchCV(estimator=LinearSVC(dual=False),
                           param_grid=param_grid,
                           scoring=scoring,
                           refit='f1',
                           return_train_score=True,
                           cv=7)
```

```python
if not hasattr(grid_clf_cv, 'cv_results_'):
    grid_clf_cv.fit(X_vec, y_train)
```

```python
coeffs = [tup['C'] for tup in grid_clf_cv.cv_results_['params']]
weights = [tup['class_weight'] for tup in grid_clf_cv.cv_results_['params']]
df_cv_result = pd.DataFrame({'C': coeffs,
                             'weights': weights,
                             'recall': grid_clf_cv.cv_results_['mean_test_recall'],
                             'precision': grid_clf_cv.cv_results_['mean_test_precision'],
                             'mean_test_f1': grid_clf_cv.cv_results_['mean_test_f1'],
                             'std_test_f1': grid_clf_cv.cv_results_['std_test_f1'],
                             'mean_test_accuracy': grid_clf_cv.cv_results_['mean_test_accuracy'],
                             'mean_test_balanced_accuracy': grid_clf_cv.cv_results_['mean_test_balanced_accuracy'],
                             'mean_time': grid_clf_cv.cv_results_['mean_score_time'],
                             'std_time': grid_clf_cv.cv_results_['std_score_time'],
                            })

df_cv_result.loc[::2, 'weights'] = 'not'
df_cv_result.round(3)
```


Имбаланс в датасете дал о себе знать и сбалансированный классификатор в целом вел себя гораздо лучше для _recall_, что достаточно очевидно. _precision_ ожидаемо падал с уменьшением параметра регуляризации, и в целом работал лучше на несбалансированных весах. Осталось найти оптимальное их отношение в _f1 score_, а так же учесть время тренировки.
Так же были рассчитаны различные средние отклонения для соблюдения bias-variance trade-off. Но они оказались все довольно одинаковыми.

```python
# balanced and not
plt.figure(figsize=(10, 7))

sns.barplot(x=df_cv_result.C.astype(str), y=df_cv_result.mean_test_f1, hue=df_cv_result.weights, alpha=0.85)

plt.ylim(0.960, 0.995)
plt.legend(facecolor='w')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax = np.ravel(ax)


sns.barplot(x=df_cv_result.C.astype(str), y=df_cv_result.recall, hue=df_cv_result.weights, alpha=0.8, ax=ax[0], zorder=1)
sns.barplot(x=df_cv_result.C.astype(str), y=df_cv_result.precision, hue=df_cv_result.weights, alpha=0.8, ax=ax[1])

sns.scatterplot(x=df_cv_result.C.astype(str), y=df_cv_result.mean_test_accuracy, ax=ax[0], color=sns.color_palette()[2], zorder=2)
sns.lineplot(x=df_cv_result.C.astype(str), y=df_cv_result.mean_test_accuracy, ax=ax[0], color=sns.color_palette()[2], zorder=2)

sns.scatterplot(x=df_cv_result.C.astype(str), y=df_cv_result.mean_test_balanced_accuracy, ax=ax[1], color=sns.color_palette()[2], zorder=2)
sns.lineplot(x=df_cv_result.C.astype(str), y=df_cv_result.mean_test_balanced_accuracy, ax=ax[1], color=sns.color_palette()[2], zorder=2)

ax[0].set_title('recall')
ax[1].set_title('precision')

ax[0].set_ylim(0.78, 1.02)
ax[1].set_ylim(0.98, 0.999)

ax[0].legend(facecolor='w')
ax[1].legend(facecolor='w')
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

balanced_df = df_cv_result[df_cv_result.weights == 'balanced']
#sns.barplot(x=balanced_df.C.astype(str), y=balanced_df.mean_test_f1, color=sns.color_palette()[3], alpha=0.8, ax=ax)

sns.scatterplot(x=balanced_df.C.astype(str), y=balanced_df.recall, ax=ax, label='recall')
sns.scatterplot(x=balanced_df.C.astype(str), y=balanced_df.precision, ax=ax, label='precision')

sns.lineplot(x=balanced_df.C.astype(str), y=balanced_df.recall, ax=ax)
sns.lineplot(x=balanced_df.C.astype(str), y=balanced_df.precision, ax=ax)

sns.scatterplot(x=balanced_df.C.astype(str), y=balanced_df.mean_test_f1, ax=ax, label='f1 score')
sns.lineplot(x=balanced_df.C.astype(str), y=balanced_df.mean_test_f1, ax=ax)

plt.fill_between(balanced_df.C.astype(str), balanced_df.mean_test_f1 + balanced_df.std_test_f1,
                 balanced_df.mean_test_f1 - balanced_df.std_test_f1, color=sns.color_palette()[2], alpha=0.3)

#plt.ylim(0.960, 1)
plt.legend(facecolor='w')
```

С таким подбором параметров не удивительно, что _recall_ оказался в большем приоритете у модели. Тем не менее, конечно, логично ожидать, что из-за того, что трейновый датасет был создал из сообщений тех же чатов, на которых была обучена модель, мы не могли не избежать переобучения. Поэтому, конечно, стоит ожидать падение скора на новых неизвестных данных.

# Процесс калибровки:

Калибровка модели SVM с использованием логистической регрессии происходит в несколько этапов:

1. **Обучение SVM**: Сначала обучается модель SVM на тренировочных данных. Этот шаг был исследован выше.

2. **Применение калибровочного классификатора**: Результаты, полученные от SVM, используются как входные данные для калибровочного классификатора, внутри которого реализована логистическая регрессия. Это позволяет преобразовать выходные данные SVM в вероятности, более интерпретируемые для конечного пользователя.

3. **Обучение логистической регрессии**: Логистическая регрессия обучается на основе выходных данных SVM, адаптируя свои параметры таким образом, чтобы лучше соответствовать исходным меткам классов. В результате получается модель, способная выдавать вероятности принадлежности к каждому из классов.

## Значение калибровки

Так как калибровка модели SVM с использованием логистической регрессии имеет ряд преимуществ:

- **Улучшенная интерпретируемость**: Преобразование маргинальных расстояний SVM в вероятности делает результаты модели более понятными для анализа и принятия решений.
- **Повышение точности классификации**: Калибровка может улучшить общую точность классификации за счет более точного присвоения вероятностей классам.

Применение калибровочного классификатора с логистической регрессией к результатам SVM позволяет получить модель, которая сочетает в себе преимущества обеих подходов, обеспечивая высокую точность и улучшенную интерпретируемость результатов классификации.

Еще одной мотивацией было исследование результатов с помощью калибровочной прямой для осознания того, как она работает на данных с разным натуральным распределением мата.

```python
from sklearn.calibration import CalibratedClassifierCV

calibrated_clf = CalibratedClassifierCV(base_estimator=LinearSVC(**grid_clf_cv.best_params_))
calibrated_clf.fit(X_vec, y_train)
```

```python
y_pred = calibrated_clf.predict_proba(X_vec_test)
calibrated_clf.predict(X_vec_test)
```

```python
from sklearn.calibration import calibration_curve

plt.figure(figsize=(10, 7))
plt.title('calibration curve')
true_probs, pred_probs = calibration_curve(y_test, y_pred[:, 1], n_bins=220, strategy='quantile')
sns.scatterplot(x=pred_probs, y=true_probs, label='mine callibration')
sns.lineplot(x=pred_probs, y=true_probs)

sns.scatterplot(x=pred_probs, y=pred_probs, label='perfect callibration')
sns.lineplot(x=pred_probs, y=pred_probs)

plt.legend(facecolor='w')
```

### Результат калибровочной модели

Мы получили то, что ожидали: неуверенный (_underconfident_) классификатор, на вероятностях, близких к 1 или 0. SVM модели в принципе подвержены проблемы неуверенности из-за природы самого алгоритма: обращая внимания на трудноразделимые точки у границы гиперплоскости мы можем "отнять" вероятность у вполне хорошо классифицируемых точек.

```python
print(classification_report(calibrated_clf.predict(X_vec_test), y_test))
```
